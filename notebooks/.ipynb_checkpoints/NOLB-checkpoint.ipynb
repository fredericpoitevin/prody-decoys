```python
%matplotlib inline
from matplotlib import pyplot as plt
import os, wget, subprocess
```

```python
pdb_id = '4AKE'
local_storage = './data/'
```

```python
if not os.path.isfile(f'{local_storage}{pdb_id.lower()}.pdb'):
    pdb_url=f'https://files.rcsb.org/download/{pdb_id.upper()}.pdb'
    pdb_filename = wget.download(pdb_url, out='./data/')
```

# NOLB

```python
n_decoys=100
rmsd=1
```

```python
run = subprocess.run(['NOLB', 
                      f'{local_storage}{pdb_id.lower()}.pdb', 
                      f'-s {n_decoys}',
                      f'--rmsd {rmsd}'
                     ], 
                     stdout=subprocess.PIPE, stderr=subprocess.PIPE
                    )
```

```python
print(run.stdout.decode('ascii'))
```
